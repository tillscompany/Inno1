**Работа с каналами**

```python
img=imread('tiger.png')
imshow(img);
```

```python
img_f=img_as_float(img)
```

**Работа с каналами**

```python
r=img_f[:,:,0]
g=img_f[:,:,1]
b=img_f[:,:,2]
imshow(r);
```

```python
from numpy import dstack
img_combined=dstack((r,b,g))
imshow(img_combined);
```

```python
avg=(g+r+b)/3
imshow(avg);
```

```python
avg=0.299*r+0.587*g+0.114*b
imshow(avg);
```
